1)What is the proportion of students placed versus not placed?

2)What is the average salary of placed students?

```python
df = pd.read_csv("D:\\DATA SCIENCE\\PROJECT\\Placement_Data_Full_Class.csv")

# Filter data for only placed students
placed_students = df[df['status'] == 'Placed']

# Calculate the average salary of placed students
average_salary = placed_students['salary'].mean()

# Plotting the average salary
plt.figure(figsize=(6, 4))
plt.bar(['Average Salary'], [average_salary], color='skyblue')
plt.ylabel('Salary (in INR)')
plt.title('Average Salary of Placed Students')
plt.ylim(0, max(placed_students['salary']) * 1.1)
plt.text(0, average_salary + 5000, f'{average_salary:.2f}', ha='center')
plt.show()
```

14)Do students with Science & Technology degrees have higher placement rates compared to those in Commerce & Management?

```python
# Filter the data for students with Science & Technology and Commerce & Management degrees
science_tech_students = df[df['degree_t'] == 'Sci&Tech']
commerce_management_students = df[df['degree_t'] == 'Comm&Mgmt']

# Calculate the placement rate for Science & Technology students
science_tech_placement_rate = science_tech_students['status'].value_counts(normalize=True)['Placed']

# Calculate the placement rate for Commerce & Management students
commerce_management_placement_rate = commerce_management_students['status'].value_counts(normalize=True)['Placed']

# Display the results
print(f"Placement rate for Science & Technology students: {science_tech_placement_rate * 100:.2f}%")
print(f"Placement rate for Commerce & Management students: {commerce_management_placement_rate * 100:.2f}%")
```

```python
# Filter the data for students with Science & Technology and Commerce & Management degrees
science_tech_students = df[df['degree_t'] == 'Sci&Tech']
commerce_management_students = df[df['degree_t'] == 'Comm&Mgmt']

# Calculate the placement rate for Science & Technology students
science_tech_placement_rate = science_tech_students['status'].value_counts(normalize=True)['Placed']

# Calculate the placement rate for Commerce & Management students
commerce_management_placement_rate = commerce_management_students['status'].value_counts(normalize=True)['Placed']

# Prepare the data for plotting
degree_types = ['Science & Technology', 'Commerce & Management']
placement_rates = [science_tech_placement_rate, commerce_management_placement_rate]

# Plot a bar chart
plt.figure(figsize=(6,4))
plt.bar(degree_types, placement_rates, color=['lightblue', 'lightcoral'])
plt.title('Placement Rates by Degree Type')
plt.xlabel('Degree Type')
plt.ylabel('Placement Rate')
plt.ylim(0, 1)  # The placement rate is a proportion, so it ranges from 0 to 1
plt.tight_layout()
plt.show()
```

15)Are there any students with high academic performance who are not placed? Why might this be the case?

```python
# Define a threshold for "high academic performance" (e.g., top 25% of students in terms of MBA percentage)
high_academic_threshold = df['mba_p'].quantile(0.75)

# Filter the data for students with high academic performance (e.g., MBA percentage above the 75th percentile)
high_academic_students = df[df['mba_p'] >= high_academic_threshold]

# Further filter to find students who are not placed
high_academic_not_placed = high_academic_students[high_academic_students['status'] == 'Not Placed']

# Count the number of such students
num_high_academic_not_placed = high_academic_not_placed.shape[0]

# Display the result
print(f"Number of high academic students who are not placed: {num_high_academic_not_placed}")
print(f"Details of students: \n{high_academic_not_placed[['mba_p', 'ssc_p', 'hsc_p', 'status']]}")
```

```python
# Define a threshold for "high academic performance" (e.g., top 25% of students in terms of MBA percentage)
high_academic_threshold = df['mba_p'].quantile(0.75)

# Filter the data for students with high academic performance (e.g., MBA percentage above the 75th percentile)
high_academic_students = df[df['mba_p'] >= high_academic_threshold]

# Further filter to find students who are placed and not placed
placement_status = high_academic_students['status'].value_counts()

# Plot a bar chart
plt.figure(figsize=(6,4))
placement_status.plot(kind='bar', color=['lightgreen', 'lightpink'])
plt.title('Placement Status of High Academic Performance Students')
plt.xlabel('Placement Status')
plt.ylabel('Number of Students')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
